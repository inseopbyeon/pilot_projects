```python
def fib_rec(num):
    if num <= 1:
        return num
    else:
        return fib_rec(num-1) + fib_rec(num-2)
    
```

```python
def fact_rec(num):
    if num<1:
        return 1
    else:
        print('rec: {}, {}'.format(num,num-1))
        return num*fact_rec(num-1)

start=time.time()
print(fact_rec(13))
end=time.time()
print(end-start)
```

```python
query = """
    SELECT 
        SUM(d.Quantity) "Count", 
        SUM(d.Quantity * p.Price) "Sales", 
        ROUND(AVG(d.Quantity * p.Price), 2) "avg",
        SUBSTR(o.OrderDate, 0, 8) "month" 
    FROM 
        OrderDetails d
        JOIN
            Products p
            ON p.ProductID = d.ProductID
        JOIN
            Orders o
            ON d.OrderID = o.OrderID
    GROUP BY
        substr(o.OrderDate, 0, 8)
    HAVING
        d.Quantity >= 20
    ;
"""
pd.read_sql(query, db)
```

```python
def gist_list():
    results = ch_driver.find_elements_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a')
    posting=[]
    
    for article in results:

        address = article.get_attribute('href')   

        id_subject = article.find_element_by_class_name('subject').text.split(']')
        post_id = int(id_subject[0].strip('['))
        subject = id_subject[1].strip(' ')

        date_cat = article.find_element_by_class_name('date').text.split(' ');date_cat
        date_time = (date_cat[0],date_cat[1])
        cat = date_cat[2]

        posting.append((post_id, date_time, cat, subject, address))
        
    return posting
```

```python
ch_driver.implicitly_wait(5)

end_result=[]
for i in range(1,10+1):
    ch_driver.execute_script('fnPagingMove({page});'.format(page=i))
    sleep(1)
    end_result.extend(gist_list())

end_result
```
